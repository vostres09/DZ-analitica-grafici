Условие 2: Задача 2
2.1. Изучите распределение домов от наличия вида на набережную

```python
# Посмотрим на распределение домов с видом на набережную
waterfront = houses['waterfront'].value_counts()
waterfront.plot(kind='bar')
plt.title('Распределение домов с видом на набережную')
plt.xlabel('Наличие вида на набережную (0 - нет, 1 - да)')
plt.ylabel('Количество объектов')
plt.show()
```

Вывод: Большинство домов в данном наборе не имеют вида на набережную. Только небольшое количество домов имеют вид на набережную.

2.2. Изучите распределение этажей домов

```python
# Посмотрим на распределение этажей домов
houses['floors'].hist(bins=10)
plt.title('Распределение этажей домов')
plt.xlabel('Количество этажей')
plt.ylabel('Количество объектов')
plt.show()
```

Вывод: Распределение этажей домов показывает, что большинство домов имеют 1 или 2 этажа. Количество домов с большим количеством этажей быстро уменьшается.

2.3. Изучите распределение состояния домов

```python
# Посмотрим на распределение состояния домов
condition_counts = houses['condition'].value_counts().sort_index()
condition_counts.plot(kind='bar')
plt.title('Распределение состояния домов')
plt.xlabel('Состояние дома')
plt.ylabel('Количество объектов')
plt.xticks(rotation=45)
plt.show()
```

Вывод: На графике видно, что большинство домов находятся в хорошем или среднем состоянии (3 или 4 по шкале). Крайние значения (1 - плохое состояние и 5 - отличное состояние) встречаются реже.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='ocean_proximity', y='median_house_value', data=houses)
plt.xticks(rotation=45)
plt.title('Влияние близости к океану на стоимость')
plt.xlabel('Близость к океану')
plt.ylabel('Медианная стоимость дома')
plt.show()
```
